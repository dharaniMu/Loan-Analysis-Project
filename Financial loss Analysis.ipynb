```python
def func(loandata):
    loandata.loc[(loandata["annual_inc"] >= 4000) & (loandata["annual_inc"] <=50000), "annual_inc_range"] = "very low"
    loandata.loc[(loandata["annual_inc"] > 50000) & (loandata["annual_inc"] <=100000), "annual_inc_range"] = "low"
    loandata.loc[(loandata["annual_inc"] > 100000) & (loandata["annual_inc"] <= 1000000), "annual_inc_range"] = "medium"
    loandata.loc[(loandata["annual_inc"] > 1000000) & (loandata["annual_inc"] <=6000000), "annual_inc_range"] = "high"
    
    return loandata

result = func(loandata)
```
